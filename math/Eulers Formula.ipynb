```python
import numpy as np
from scipy.fft import fft, ifft, fftfreq

%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
SF = 2.5
# start of observed time
T1 = 0
# end of observed time
T2 = 5
# sampling frequency
F = 1.0 / 80
N = int((T2-T1)/F)
time_sample = np.linspace(T1, T2, N, endpoint=False)
y = np.exp(-2j * np.pi * time_sample * SF)
```

```python
plt.figure(figsize=(6, 6), dpi=80)
plt.plot(time_sample, y.real)
plt.plot(time_sample, y.imag)
```

```python
plt.figure(figsize=(6, 6), dpi=80)
plt.plot(y.real, y.imag)
```

```python
fig = plt.figure(figsize=(6, 6), dpi=80)
ax = fig.add_subplot(111)
plt.ion()

fig.show()
fig.canvas.draw()

for i in range(time_sample.shape[0]):
    #ax.clear()
    ax.scatter(y.real[i], y.imag[i], color="red", alpha=0.2)
    fig.canvas.draw()
```
