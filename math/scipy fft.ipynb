```python
import numpy as np
from scipy.fft import fft, ifft, fftfreq
import matplotlib.pyplot as plt
```

```python
SF = 2
# start of observed time
T1 = 0
# end of observed time
T2 = 5
# sampling frequency
F = 1.0 / 80
# number of samples
N = int((T2-T1)/F)
time_sample = np.linspace(T1, T2, N, endpoint=False)
wave_sample = np.sin(SF * (2*np.pi) * time_sample)     # true function f(t)
```

```python
# Generate a 2 hertz sine wave that lasts for 5 seconds
plt.figure(figsize=(12, 6), dpi=80)
plt.scatter(time_sample, wave_sample);
```

#### FFFT

```python
wave_sample_ftf = fft(wave_sample)
signal_frequencies = fftfreq(N, F)
plt.figure(figsize=(12, 6), dpi=80)
#plt.bar(signal_frequencies[0:N//2], np.abs(wave_sample_ftf[0:N//2]))
plt.bar(signal_frequencies, np.abs(wave_sample_ftf))
```

```python
wave_sample_ftf = fft(wave_sample)
signal_frequencies = fftfreq(N, F)
plt.figure(figsize=(12, 6), dpi=80)
plt.bar(signal_frequencies[0:N//2], np.abs(wave_sample_ftf[0:N//2]))
#plt.bar(signal_frequencies, np.abs(wave_sample_ftf))
```

```python
plt.figure(figsize=(12, 6), dpi=80)
plt.scatter(time_sample, np.abs(ifft(wave_sample_ftf)), label="ifft abs");
plt.scatter(time_sample, np.real(ifft(wave_sample_ftf)), label="ifft real part");
plt.scatter(time_sample, np.imag(ifft(wave_sample_ftf)), label="ifft imaginary part");
plt.plot(time_sample, wave_sample, label="true function");
plt.legend();
```
