```python
import numpy as np
from scipy.fft import fft, ifft, fftfreq

%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
SF = 5
# start of observed time
T1 = 0
# end of observed time
T2 = 5
# sampling frequency
F = 1.0 / 20
# number of samples
N = int((T2-T1)/F)
time_sample = np.linspace(T1, T2, N, endpoint=False)
wave_sample = np.sin(SF * (2*np.pi) * time_sample)     # true function g(t)
```

```python
# Generate a 2 hertz sine wave that lasts for 5 seconds
plt.plot(time_sample, wave_sample);
```

#### FFT by formula

```python
n=1
wave_sample_fft = np.sin(SF * (2*np.pi) * time_sample) * np.exp(n * (-2j * np.pi * F * time_sample))
(np.round(wave_sample_fft.real.sum(), 2), np.round(wave_sample_fft.imag.sum(), 2))
```

```python
np.round(np.abs(wave_sample_fft),2).sum()
```

```python
n=2
wave_sample_fft = np.sin(SF * (2*np.pi) * time_sample) * np.exp(n * (-2j * np.pi) * time_sample)/(T2-T1)
(np.round(wave_sample_fft.real.sum(), 2), np.round(wave_sample_fft.imag.sum(), 2))
```

```python
np.abs(wave_sample_fft).mean()
```

```python
plt.figure(figsize=(6, 6), dpi=80)
plt.plot(time_sample, wave_sample_fft.real)
plt.plot(time_sample, wave_sample_fft.imag)
```

```python
plt.figure(figsize=(6, 6), dpi=80)
plt.plot(wave_sample_fft.real, wave_sample_fft.imag)
```

```python
fig = plt.figure(figsize=(6, 6), dpi=80)
ax = fig.add_subplot(111)
plt.ion()

fig.show()
fig.canvas.draw()

for i in range(time_sample.shape[0]):
    #ax.clear()
    ax.scatter(wave_sample_fft.real[i], wave_sample_fft.imag[i], color="red", alpha=0.2)
    fig.canvas.draw()
```

```python
import numpy as np
from matplotlib import pyplot as plt

SAMPLE_RATE = 44100  # Hertz
DURATION = 5  # Seconds

def generate_sine_wave(freq, sample_rate, duration):
    x = np.linspace(0, duration, sample_rate * duration, endpoint=False)
    frequencies = x * freq
    # 2pi because np.sin takes radians
    y = np.sin((2 * np.pi) * frequencies)
    return x, y

# Generate a 2 hertz sine wave that lasts for 5 seconds
x, y = generate_sine_wave(2, SAMPLE_RATE, DURATION)
plt.plot(x, y)
plt.show()
```
